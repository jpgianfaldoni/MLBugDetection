## Random Forest

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("Datasets/fraud-data.csv")
```

```python
data.head()
```

```python
data.nameOrig = data.nameOrig.str.replace("C", "")
data.nameOrig = data.nameOrig.str.replace("M", "")
data.nameDest = data.nameDest.str.replace("C", "")
data.nameDest = data.nameDest.str.replace("M", "")
```

```python
data.head()
```

```python
data["type"].unique()
```

```python
encoder = OneHotEncoder()
feature = encoder.fit_transform(data[["type"]]).toarray()
```

```python
encoder.categories_
```

```python
feature_labels = encoder.categories_
feature_labels = np.array(feature_labels).ravel()
```

```python
print(feature_labels)
```

```python
featuresAll = pd.DataFrame(feature, columns = feature_labels)
```

```python
data_new = pd.concat([data, featuresAll], axis=1)
```

```python
data_new
```

```python
x = data.drop(["isFraud", "type"], axis=1)
y = data.isFraud

xTrain, xTest, yTrain, yTest = train_test_split(x, y, test_size=.99)
```

```python
xTrain.shape
```

```python
clf = RandomForestClassifier()
```

```python
clf.fit(xTrain, yTrain)
```

```python
y_pred = clf.predict(xTest)
```

```python
from sklearn import metrics

print("Accuracy:",metrics.accuracy_score(yTest, y_pred))
```
